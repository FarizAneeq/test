##  Adding Limits

```python
import seaborn as sns
import matplotlib.pyplot as plt

phool= sns.load_dataset("iris")
phool

#draw lineplot

sns.lineplot(x="sepal_length", y="sepal_width" , data = phool)
plt.title("Phoolon the Plot")
plt.xlim(2)

plt.ylim(1.5)
plt.show()
```

## Size of Figure

```python
import seaborn as sns
import matplotlib.pyplot as plt

phool= sns.load_dataset("iris")
sns.set_style("white")

plt.figure(figsize=(10,8))

#draw lineplot


sns.lineplot(x="sepal_length", y="sepal_width" , data = phool)
plt.title("Phoolon the Plot")


plt.show()
```
